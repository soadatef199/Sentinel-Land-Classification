```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# تحليل عدد الصور في كل مجموعة
def analyze_image_counts(df, name):
    print(f"The number of images in the data {name}: {len(df)}")
    label_counts = df['ClassName'].value_counts()
    print(f"\nClass distribution in the data {name}:")
    print(label_counts)
    
    # رسم توزيع الفئات
    plt.figure(figsize=(10, 5))
    sns.barplot(x=label_counts.index, y=label_counts.values)
    plt.xticks(rotation=45)
    plt.xlabel("Class (Label)")
    plt.ylabel("Number of images")
    plt.title(f"Distribution of the number of images per class in the data {name}")
    plt.show()

# تحليل كل الملفات
analyze_image_counts(train_df, "Trainning")
analyze_image_counts(test_df, "Testing")
analyze_image_counts(val_df, "Validating")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# دالة لتحليل الفئات (Labels)
def analyze_labels(df, name):
    # حساب عدد الصور لكل فئة
    label_counts = df['ClassName'].value_counts()
    label_percentages = (label_counts / label_counts.sum()) * 100

    # طباعة النتائج
    print(f"\nClass analysis in the data {name}:")
    print(label_counts)
    print("\nPercentage of each class:")
    print(label_percentages)

    # رسم توزيع الفئات (Bar Chart)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=label_counts.index, y=label_counts.values, palette="viridis")
    plt.xticks(rotation=45)
    plt.xlabel("Class (Label)")
    plt.ylabel("Number of images")
    plt.title(f"Distribution of the number of images per class in the data {name}")
    plt.show()

    # رسم النسب المئوية (Pie Chart) توضح توازن البيانات
    plt.figure(figsize=(8, 8))
    plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", len(label_counts)))
    plt.title(f"Percentage of each class in the data {name}")
    plt.show()

# تحليل الفئات في كل مجموعة بيانات
analyze_labels(train_df, "Trainning")
analyze_labels(test_df, "Testing")
analyze_labels(val_df, "Validating")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# المجلد الرئيسي اللي فيه الصور
base_folder = r"C:\Users\Amany\Depi\Senital-2\archive\EuroSAT"

# القنوات اللونية لجمع البيانات
r_channel = []
g_channel = []
b_channel = []

# نمر على كل الفولدرات داخل المجلد الرئيسي
for class_folder in os.listdir(base_folder):
    class_path = os.path.join(base_folder, class_folder)
    
    # نتأكد إنه فولدر (عشان ما نقرأش ملفات غير الصور بالغلط)
    if os.path.isdir(class_path):
        print(f"Processing: {class_folder}")
        
        # نمر على كل الصور في الفولدر
        for image_name in os.listdir(class_path):
            image_path = os.path.join(class_path, image_name)
            
            try:
                # قراءة الصورة
                image = Image.open(image_path)
                image_array = np.array(image)
                
                # فصل القنوات اللونية
                r_channel.extend(image_array[:, :, 0].flatten())
                g_channel.extend(image_array[:, :, 1].flatten())
                b_channel.extend(image_array[:, :, 2].flatten())
            
            except Exception as e:
                print(f"Error with {image_name}: {e}")

# رسم التوزيع لكل قناة لونية
plt.figure(figsize=(10, 4))
plt.hist(r_channel, bins=50, color='red', alpha=0.5, label="R Channel")
plt.hist(g_channel, bins=50, color='green', alpha=0.5, label="G Channel")
plt.hist(b_channel, bins=50, color='blue', alpha=0.5, label="B Channel")

plt.title("Spectral Band Distribution (RGB)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Assuming band 4 = Red, band 3 = Green, band 2 = Blue
rgb_image = np.stack([image[3], image[2], image[1]], axis=-1)

# Normalize for display
rgb_image = (rgb_image - np.min(rgb_image)) / (np.max(rgb_image) - np.min(rgb_image))

plt.figure(figsize=(6, 6))
plt.imshow(rgb_image)
plt.title('RGB Image')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot histogram for the first 4 bands
plt.figure(figsize=(12, 8))
for i in range(4):
    plt.hist(image[i].flatten(), bins=50, alpha=0.5, label=f'Band {i+1}')

plt.legend()
plt.title("Pixel Value Distribution")
plt.show()
```

```python
import seaborn as sns

# Flatten each band and stack them into a matrix
bands_flattened = np.array([image[i].flatten() for i in range(image.shape[0])]).T

# Create a correlation matrix
correlation = np.corrcoef(bands_flattened, rowvar=False)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, cmap='coolwarm', annot=True, fmt='.2f', xticklabels=[f'Band {i+1}' for i in range(image.shape[0])], yticklabels=[f'Band {i+1}' for i in range(image.shape[0])])
plt.title('Correlation Between Bands')
plt.show()
```

```python
# Extract Red and NIR bands
red = image[3]
nir = image[7]

# NDVI calculation
ndvi = (nir - red) / (nir + red)

# Plot NDVI
plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar()
plt.title('NDVI')
plt.show()

# NDVI statistics
print(f"NDVI Mean: {np.mean(ndvi):.2f}")
print(f"NDVI Std: {np.std(ndvi):.2f}")
```

```python
import os
import matplotlib.pyplot as plt
import cv2
import random
import numpy as np

main_folder = r"D:\final_ project\2750"


class_images = {}
for class_name in os.listdir(main_folder):
    class_path = os.path.join(main_folder, class_name)
    
    if os.path.isdir(class_path):
        images = [
            os.path.join(class_path, file)
            for file in os.listdir(class_path)
            if file.lower().endswith(('.png', '.jpg', '.jpeg'))
        ]
        if images:
            class_images[class_name] = images


if not class_images:
    print("No images")
else:
    print(f"There ara classes {len(class_images)}")


num_classes_to_display = min(10, len(class_images))  
selected_classes = random.sample(list(class_images.keys()), num_classes_to_display)


plt.figure(figsize=(15, 4 * num_classes_to_display))

for class_idx, class_name in enumerate(selected_classes):
    images = class_images[class_name]
    sampled_images = random.sample(images, min(5, len(images)))  

    
    hist_r, hist_g, hist_b = np.zeros(256), np.zeros(256), np.zeros(256)
    
    for img_path in sampled_images:
        img = cv2.imread(img_path)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    
        for i, (hist, color) in enumerate(zip([hist_r, hist_g, hist_b], ['r', 'g', 'b'])):
            hist += cv2.calcHist([img_rgb], [i], None, [256], [0, 256]).flatten()

    
    hist_r /= hist_r.max()
    hist_g /= hist_g.max()
    hist_b /= hist_b.max()


    plt.subplot(num_classes_to_display, 1, class_idx + 1)
    plt.plot(hist_r, color='r', label="Red")
    plt.plot(hist_g, color='g', label="Green")
    plt.plot(hist_b, color='b', label="Blue")
    plt.title(f"class: {class_name}", fontsize=12, fontweight='bold')
    plt.xlabel("Color Intensity")
    plt.ylabel("اNormalized Frequency")
    plt.legend()

plt.tight_layout()
plt.show()
```
